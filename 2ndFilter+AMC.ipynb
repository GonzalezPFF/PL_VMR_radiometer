```python
import pandas as pd
import numpy as np
structured_df = pd.read_csv('/content/drive/My Drive/NICOS_NN/SOLSTICIO2_0/merradf_output_structured.csv')
structured_df.head()
```

```python
def safe_eval_and_convert(cell):
    try:
        evaluated = eval(cell, {"np": np})
        return [float(x) for x in evaluated]
    except Exception as e:
        return cell
```

```python
structured_df = structured_df.applymap(safe_eval_and_convert)
structured_df.head()
```

**For do the separation in type of layer; finding the nearest averaged-index closest to 225 mbar to tropopause; and 1 mbar for stratopause**

```python
tropopause_indices = []
stratopause_indices = []
for idx, row in structured_df.iterrows():
    pressures = np.array(row[' PL[mbar]'])

    tropopause_idx = np.abs(pressures - 225).argmin()
    tropopause_indices.append(tropopause_idx)

    stratopause_idx = np.abs(pressures - 1).argmin()
    stratopause_indices.append(stratopause_idx)

avg_tropopause_idx = int(round(np.mean(tropopause_indices)))
avg_stratopause_idx = int(round(np.mean(stratopause_indices)))
```

```python
print(f"Average Tropopause Index (closest to 225 mbar): {avg_tropopause_idx}")
print(f"Average Stratopause Index (closest to 1 mbar): {avg_stratopause_idx}")
```

```python
import os
output_dir = '/content/drive/MyDrive/NICOS_NN/SOLSTICIO2_0/AMC_files/'
os.makedirs(output_dir, exist_ok=True)
```

```python
MESOSPHERE_END = avg_stratopause_idx
STRATOSPHERE_END = avg_tropopause_idx
TOTAL_LAYERS = 67
```

```python
import sys
import os

for idx, row in structured_df.iterrows():
    pressures = row[' PL[mbar]']
    h2o = row[' H2O[vmr]']
    temps = row[' T[K]']
    delp = row[' DELP[mbar]']

    output_file = os.path.join(output_dir, f"{idx+1}.amc")
    print(f"Creating file: {idx+1} of {len(structured_df)}")

    with open(output_file, 'w') as f:
        original_stdout = sys.stdout
        sys.stdout = f

        try:
            # Write the header (same as original)
            print("?")
            print("? Usage:")
            print("?   am this_file f_min f_max df zenith_angle trop_h2o_scale_factor")
            print("? Example:")
            print("?   am this_file 20 GHz  26 GHz  60 KHz  0 deg  1.0")
            print("?")
            print("f %1 %2  %3 %4 %5 %6")
            print("output f GHz  Tb K Trj K")
            print("za %7 %8")
            print("tol 1e-4")
            print("")
            print("Nscale troposphere h2o %9")
            print("")
            print("T0 2.7 K")

            # Write mesosphere data (layers 0-7)
            for n in range(MESOSPHERE_END):
                print('')
                print('layer mesosphere')
                print(f'Pbase {pressures[n]:.10e} mbar')
                print(f'Tbase {temps[n]:.10e} K')
                print('column dry_air vmr')
                print(f'column h2o vmr {h2o[n]:.10e}')

            # Write stratosphere data (layers 8-35)
            for n in range(MESOSPHERE_END, STRATOSPHERE_END):
                print('')
                print('layer stratosphere')
                print(f'Pbase {pressures[n]:.10e} mbar')
                print(f'Tbase {temps[n]:.10e} K')
                print('lineshape Voigt-Kielkopf')
                print('column dry_air vmr')
                print(f'column h2o vmr {h2o[n]:.10e}')

            # Write troposphere data (layers 36-66)
            for n in range(STRATOSPHERE_END, TOTAL_LAYERS):
                print('')
                print('layer troposphere')
                print(f'Pbase {pressures[n]:.10e} mbar')
                print(f'Tbase {temps[n]:.10e} K')
                print('column dry_air vmr')
                print(f'column h2o vmr {h2o[n]:.10e}')

        finally:
            sys.stdout = original_stdout

print("Processing complete.")
```
