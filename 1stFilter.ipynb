```python
import pandas as pd

df = pd.read_csv(filepath, delimiter=',', header=0)
```

```python
df
```

```python
chunk_size = 67
structured_data = []
for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i+chunk_size]
    chunk_arrays = [chunk[col].to_numpy() for col in df.columns]
    structured_data.append(chunk_arrays)
```

```python
import numpy as np
structured_df = pd.DataFrame(structured_data, columns=df.columns)
```

```python
import matplotlib.pyplot as plt

pl_values = structured_df.iloc[0][2]
h2o_values = structured_df.iloc[0][3]

if isinstance(pl_values, np.ndarray):
    pl_values = pl_values.tolist()

if isinstance(h2o_values, np.ndarray):
    h2o_values = h2o_values.tolist()

plt.plot( h2o_values, pl_values)
plt.ylabel('PL[mbar]')
plt.xlabel('H2O[vmr]')
plt.gca().invert_yaxis()
plt.title('Pressure Layer (interp.) vs H2O m/m from obj1')
plt.show()
```
